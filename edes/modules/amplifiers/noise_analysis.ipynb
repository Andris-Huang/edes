```python
from edes.modules.detection.detection_utils import plot, plot_ax, big_plt_font, dB_to_frac, frac_to_dB, big_plt_font
import matplotlib.pyplot as plt 
import numpy as np
from scipy.constants import k
big_plt_font()
```

```python
def NF_to_Vrms(NF, T=303, Rs=50):
    F = 10**(NF/10)
    Nin = 4*k*T*Rs
    Namp = (F-1)*Nin
    return np.sqrt(Namp)*1e9
```

```python
# gain = 23 # dB
NF = 0.6 # dB
NF_to_Vrms(NF)
```

```python
NF = [0.63, 1.884, 2.701, 3.390] 
Tn = [306.13, 200, 100, 4] 
Vrms = [NF_to_Vrms(NF[i], Tn[i]) for i in range(4)] 
```

```python
plot(Tn, Vrms, 'o--', xlabel='Source noise temperature (K)', ylabel=r'$\bar{V}_{\rm{n, in}}$ (nV/$\sqrt{\text{Hz}}$')
```
