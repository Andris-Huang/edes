```python
def time_since_start(df, name='date'): 
    full_name = pd.to_datetime(df[name], format="%Y/%m/%d %H:%M:%S.%f")
    start_time = full_name.iloc[0]
    return (full_name - start_time).dt.total_seconds()
```

```python
def plot_pressure_vs_mass(file, vlines=[16, 18, 28, 44]):
    df = pd.read_csv(file, keep_default_na=True, skiprows=256, names=['date', 'mass', 'pressure', 'unknown'])
    # Round 'mass' to the nearest 0.5
    df['mass_rounded'] = df['mass'].round(1)  # ensure it's float
    df['mass_rounded'] = (df['mass_rounded'] * 2).round().div(2)
    
    # Group by rounded mass and compute average pressure
    grouped = df.groupby('mass_rounded')['pressure'].mean().reset_index()
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(grouped['mass_rounded'], grouped['pressure'], width=0.4)
    for i in range(len(vlines)): 
        plt.axvline(vlines[i], color=f'C{i+1}', alpha=0.7, linestyle='--', label=f'm = {vlines[i]}')
    plt.xlabel('Mass (rounded to nearest 0.5)')
    plt.ylabel('Average Pressure (torr)')
    plt.title('Average Pressure vs Rounded Mass')
    plt.yscale('log')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

def get_pressure_at_mass(file, m=18): 
    df = pd.read_csv(file, keep_default_na=True, skiprows=256, names=['date', 'mass', 'pressure', 'unknown'])
    # Round 'mass' to the nearest 0.5
    df['mass_rounded'] = df['mass'].round(1)  # ensure it's float
    df['mass_rounded'] = (df['mass_rounded'] * 2).round().div(2)
    
    # Group by rounded mass and compute average pressure
    grouped = df.groupby('mass_rounded')['pressure'].mean().reset_index()
    return np.array(grouped.loc[grouped['mass_rounded'] == m]['pressure'])[0]
```
