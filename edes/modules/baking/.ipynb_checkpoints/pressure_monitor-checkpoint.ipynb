```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from edes.utils.plotting import plot_dot_dashed, plot_ax_dot_dashed, plot, plot_ax, big_plt_font
from edes.utils.file_handling import list_filenames
big_plt_font()
```

```python
def time_since_start(df, name='date'): 
    full_name = pd.to_datetime(df[name], format="%Y/%m/%d %H:%M:%S.%f")
    start_time = full_name.iloc[0]
    return (full_name - start_time).dt.total_seconds()
```

```python
base = r'C:\\Users\\andrish\\Downloads/08_01_2025_electron_baking'
all_files = [f'{base}/{i}' for i in list_filenames(base)]
```

```python
def plot_pressure_vs_mass(file, vlines=[16, 18, 28, 44]):
    df = pd.read_csv(file, keep_default_na=True, skiprows=256, names=['date', 'mass', 'pressure', 'unknown'])
    # Round 'mass' to the nearest 0.5
    df['mass_rounded'] = df['mass'].round(1)  # ensure it's float
    df['mass_rounded'] = (df['mass_rounded'] * 2).round().div(2)
    
    # Group by rounded mass and compute average pressure
    grouped = df.groupby('mass_rounded')['pressure'].mean().reset_index()
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(grouped['mass_rounded'], grouped['pressure'], width=0.4)
    for i in range(len(vlines)): 
        plt.axvline(vlines[i], color=f'C{i+1}', alpha=0.7, linestyle='--', label=f'm = {vlines[i]}')
    plt.xlabel('Mass (rounded to nearest 0.5)')
    plt.ylabel('Average Pressure (torr)')
    plt.title('Average Pressure vs Rounded Mass')
    plt.yscale('log')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_ax_pressure_vs_mass(ax, file, label=None, vlines=[16, 18, 28, 44], plot_vlines=False, plot_style='.--'):
    df = pd.read_csv(file, keep_default_na=True, skiprows=256, names=['date', 'mass', 'pressure', 'unknown'])
    # Round 'mass' to the nearest 0.5
    df['mass_rounded'] = df['mass'].round(1)  # ensure it's float
    df['mass_rounded'] = (df['mass_rounded'] * 2).round().div(2)
    
    # Group by rounded mass and compute average pressure
    grouped = df.groupby('mass_rounded')['pressure'].mean().reset_index()
    
    # Plot
    if plot_style == 'bar':
        ax.bar(grouped['mass_rounded'], grouped['pressure'], width=0.4, alpha=0.7, label=label)
    else: 
        ax.plot(grouped['mass_rounded'], grouped['pressure'], plot_style, alpha=0.7, label=label)
    if plot_vlines:
        for i in range(len(vlines)): 
            plt.axvline(vlines[i], color=f'C{i+1}', alpha=0.7, linestyle='--', label=f'm = {vlines[i]}')
    ax.set_xlabel('Mass (rounded to nearest 0.5)')
    ax.set_ylabel('Average Pressure (torr)')
    ax.set_title('Average Pressure vs Rounded Mass')
    ax.set_yscale('log')
    ax.grid(True)
    ax.legend()
    plt.tight_layout()

def get_pressure_at_mass(file, m=18): 
    df = pd.read_csv(file, keep_default_na=True, skiprows=256, names=['date', 'mass', 'pressure', 'unknown'])
    # Round 'mass' to the nearest 0.5
    df['mass_rounded'] = df['mass'].round(1)  # ensure it's float
    df['mass_rounded'] = (df['mass_rounded'] * 2).round().div(2)
    
    # Group by rounded mass and compute average pressure
    grouped = df.groupby('mass_rounded')['pressure'].mean().reset_index()
    return np.array(grouped.loc[grouped['mass_rounded'] == m]['pressure'])[0]
```

```python
plot([1,2,3,4,5,7,8,9,10], [get_pressure_at_mass(f) for f in all_files[1:-1]], 'o', xlabel='Days since start', ylabel='Water pressure')
plt.yscale('log')
```
