```python
import edes 
from edes.experiments import Experiment
from edes.utils.plotting import plot, plot_ax, big_plt_font, plot_ax_errbar, plot_errbar
import numpy as np 
from tqdm import tqdm, trange 
import matplotlib.pyplot as plt 
import time
from edes.experiments.sequences.base import TipVoltageSweep
big_plt_font()
```

```python
exp = Experiment('3layer_trapping_052026')
```

```python
exp.list_devices()
```

```python
f, data = exp.SSA.measure_freq_spectrum(freq_center=198) 
```

```python
plot(f, data, alpha=1, xlabel='Frequency (MHz)', ylabel='Power (dBm)')
```

```python
V_start = 1000 
V_stop = 3000 
V_step = 100 
N_avg = 8
t_PSU_settle = 2
t_meas_delay = 0.2

all_I = [] 
all_V = np.arange(V_start, V_stop, V_step)
for V in tqdm(all_V): 
    exp.FEtip_PSU.set_voltage(V) 
    time.sleep(t_PSU_settle) 
    I_loc = []
    for _ in range(N_avg):
        I = exp.Agilent.measure_V()/100e6
        I_loc.append(I)    
        time.sleep(t_meas_delay) 
    all_I.append(I_loc) 

exp.FEtip_PSU.ramp_down(0)
```

```python
plot_errbar(all_V, -np.mean(all_I, axis=1)*1e9, yerr=np.std(all_I, axis=1)*1e9, fmt='.--', xlabel='Tip voltage (V)', ylabel='Tip current (nA)')
# plt.yscale('log')
```

```python
exp.FEtip_PSU.ramp_down(0)
```

```python
tip_Vsweep = TipVoltageSweep(saving_dir=exp.saving_dir, 
                             V_start=1000, V_stop=3000, V_step=200, R=100e6, 
                             N_avg=8, t_PSU_settle=2, t_meas_delay=0.2, 
                             FEtip_PSU=exp.FEtip_PSU, multimeter=exp.Agilent)
```

```python
file = tip_Vsweep.run_save()
```

```python
plot_errbar(file['all_V'], -np.mean(file['all_I'], axis=1)*1e12, yerr=np.std(file['all_I'], axis=1)*1e12, fmt='.--', xlabel='Tip voltage (V)', ylabel='Electrode current (pA)')
# plt.yscale('log')
```
